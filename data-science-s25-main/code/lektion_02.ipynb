# Agglomerative Hierarchisches Clustering

```python
hierarchical_engine = AgglomerativeClustering(n_clusters=3)
clusters = hierarchical_engine.fit_predict(x)
```

```python
df = data.copy()
df["clusters"] = clusters

df.plot(
    kind="scatter",
    x="description_length",
    y="words_fullname",
    c="clusters",
    cmap="inferno"
    )
```

#### wie man des ergebnis mit den anderen Daten in Beziehung setzen kann und darstellen
