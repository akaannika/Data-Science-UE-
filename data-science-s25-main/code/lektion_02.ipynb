# K-Means

```python
kmeans_engine.n_clusters
```

#### (wie viele Cluster unser kmeans engine ausgibt) = n_clusters


```python
kmeans_engine.fit(x)
```

#### .fit() = Klasse einstellen; kmeans wird trainiert -> auf info steht = fitted 

```python
kmeans_engine.predict(x)
```
